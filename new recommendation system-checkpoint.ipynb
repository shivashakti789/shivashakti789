```python
import nltk 
import nltk
nltk.download('punkt')
nltk.download('stopwords')

from nltk.tokenize import sent_tokenize, word_tokenize

EXAMPLE_TEXT = "Hello Mr.Smith,how are you doing today?"

print(sent_tokenize(EXAMPLE_TEXT))
from nltk.stem import PorterStemmer
from nltk.corpus import stopwords
```

```python
ps = PorterStemmer()
```

```python
import re 
```

```python
#function 
def clean(text):
    if isinstance(text, str):
        clean =  re.sub(r'[^a-zA-Z0-9/s]', ' ', text)
        clean = clean.lower()
        clean = nltk.word_tokenize(clean)
        stem = [ps.stem(word)for word in clean if word not in stopwords.words('english') ]
        return " ".join(stem)
    else:
        return text
```

```python
sample_df['Title'] = sample_df['Title'].astype(str).apply(lambda x: clean(x))
sample_df['Job.Description'] = sample_df['Job.Description'].astype(str).apply(lambda x: clean(x))
```

```python
sample_df['newcollumn']=sample_df['Title'].astype(str) + " " + sample_df['Job.Description'].astype(str)
```

```python
sample_df.reset_index(drop=True, inplace=True)
df = sample_df
```

```python
df
```

```python
from sklearn.feature_extraction.text  import  TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
```

```python
tfidf = TfidfVectorizer()
matrix = tfidf.fit_transform(sample_df['newcollumn'])
similarity  = cosine_similarity(matrix)
```

```python
similarity
```

```python
#recommendation 

def recommandation(Title):
    idx= sample_df[sample_df['Title'] == Title].index[0]
    idx = sample_df.index.get_loc(idx)
    distance = sorted(list(enumerate(similarity[idx])), key = lambda x:x[1], reverse = True)[1:11]
    
    jobs = []
    for i in distance:
        jobs.append(df.iloc[i[0]].Title)
    return jobs 
    
```

```python

recommandation('administr assist officeteam')
```
